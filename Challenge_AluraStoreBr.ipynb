# Gráficos
Objetivo: Analisar o dados fornecidos para calcular o faturamento ao longo do tempo e analisar os resultados de forma gráfica.

É necessário realizar algumas etapas de pré-processamento de dados para ser possível visualizá-los de forma gráfica.

```python
loja['Data da Compra'] = pd.to_datetime(loja['Data da Compra'], format='%d/%m/%Y')
loja2['Data da Compra'] = pd.to_datetime(loja2['Data da Compra'], format='%d/%m/%Y')
loja3['Data da Compra'] = pd.to_datetime(loja3['Data da Compra'], format='%d/%m/%Y')
loja4['Data da Compra'] = pd.to_datetime(loja4['Data da Compra'], format='%d/%m/%Y')
```

```python
loja['Ano'] = loja['Data da Compra'].dt.year
loja['Mes'] = loja['Data da Compra'].dt.month
loja2['Ano'] = loja2['Data da Compra'].dt.year
loja2['Mes'] = loja2['Data da Compra'].dt.month
loja3['Ano'] = loja3['Data da Compra'].dt.year
loja3['Mes'] = loja3['Data da Compra'].dt.month
loja4['Ano'] = loja4['Data da Compra'].dt.year
loja4['Mes'] = loja4['Data da Compra'].dt.month
```

```python
receita_mensal_loja1 = loja.groupby(['Ano', 'Mes'])['Preço'].sum()
receita_mensal_loja2 = loja2.groupby(['Ano', 'Mes'])['Preço'].sum()
receita_mensal_loja3 = loja3.groupby(['Ano', 'Mes'])['Preço'].sum()
receita_mensal_loja4 = loja4.groupby(['Ano', 'Mes'])['Preço'].sum()

display(receita_mensal_loja1.head())
display(receita_mensal_loja2.head())
display(receita_mensal_loja3.head())
display(receita_mensal_loja4.head())
```

# Visualizando os dados de faturamento em forma de gráfico

## Resumo:


* O faturamento mensal para cada uma das lojas foi calculado agrupando os dados por ano e mês e somando os valores da tabela 'Preço'
*O gráfico de linhas foi gerado, com o tempo sendo representado pelo eixo-x e o faturamento mensal sendo representado pelo eixo-y.
* O gráfico de colunas foi gerado, com os meses de cada ano sendo representados pelo eixo-x e o faturamento mensal representado pelo eixo-y.
* O gráfico de barras foi gerado, com o faturamento anual sendo representado pelo eixo-x e o ano sendo representado pelo eixo-y.


```python
import matplotlib.pyplot as plt
import seaborn as sns

receita_mensal_loja1 = receita_mensal_loja1.reset_index()
receita_mensal_loja2 = receita_mensal_loja2.reset_index()
receita_mensal_loja3 = receita_mensal_loja3.reset_index()
receita_mensal_loja4 = receita_mensal_loja4.reset_index()

receita_mensal_loja1['Tempo'] = receita_mensal_loja1['Ano'].astype(str) + '-' + receita_mensal_loja1['Mes'].astype(str)
receita_mensal_loja2['Tempo'] = receita_mensal_loja2['Ano'].astype(str) + '-' + receita_mensal_loja2['Mes'].astype(str)
receita_mensal_loja3['Tempo'] = receita_mensal_loja3['Ano'].astype(str) + '-' + receita_mensal_loja3['Mes'].astype(str)
receita_mensal_loja4['Tempo'] = receita_mensal_loja4['Ano'].astype(str) + '-' + receita_mensal_loja4['Mes'].astype(str)


plt.figure(figsize=(15, 8))
sns.lineplot(data=receita_mensal_loja1, x='Tempo', y='Preço', label='Loja 1')
sns.lineplot(data=receita_mensal_loja2, x='Tempo', y='Preço', label='Loja 2')
sns.lineplot(data=receita_mensal_loja3, x='Tempo', y='Preço', label='Loja 3')
sns.lineplot(data=receita_mensal_loja4, x='Tempo', y='Preço', label='Loja 4')

plt.title('Faturamento Mensal das Lojas ao Longo do Tempo')
plt.xlabel('Tempo (Ano, Mês)')
plt.ylabel('Faturamento Mensal (R$)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Concatenate the dataframes for easier plotting with seaborn
all_stores_revenue = pd.concat([
    receita_mensal_loja1.assign(Loja='Loja 1'),
    receita_mensal_loja2.assign(Loja='Loja 2'),
    receita_mensal_loja3.assign(Loja='Loja 3'),
    receita_mensal_loja4.assign(Loja='Loja 4')
])

plt.figure(figsize=(20, 10))
sns.barplot(data=all_stores_revenue, x='Tempo', y='Preço', hue='Loja')

plt.title('Faturamento Mensal das Lojas ao Longo do Tempo')
plt.xlabel('Tempo (Ano-Mês)')
plt.ylabel('Faturamento Mensal (R$)')
plt.xticks(rotation=90)
plt.legend(title='Loja')
plt.tight_layout()
plt.show()
```

#Relatório Final

Introdução

O objetivo deste relatório é realizar uma análise comparativa entre quatro lojas distintas, avaliando o desempenho de cada uma com base nos seguintes critérios: faturamento total, categorias de produtos mais e menos vendidos, média de avaliação dos clientes, produtos mais e menos vendidos e custo médio do frete.

A finalidade é identificar qual loja apresenta o pior desempenho geral e, portanto, deve ser considerada para fechamento estratégico.

Desenvolvimento

A seguir, são apresentados os principais indicadores extraídos dos dados:




```python
import pandas as pd

# Função para analisar os dados de uma loja
def analisar_loja(df, nome_loja):
    return {
        "Loja": nome_loja,
        "Faturamento Total (R$)": round(df["Preço"].sum(), 2),
        "Frete Médio (R$)": round(df["Frete"].mean(), 2),
        "Avaliação Média": round(df["Avaliação da compra"].mean(), 2),
        "Produto Mais Vendido": df["Produto"].value_counts().idxmax(),
        "Produto Menos Vendido": df["Produto"].value_counts().idxmin(),
        "Categoria Mais Vendida": df["Categoria do Produto"].value_counts().idxmax(),
        "Categoria Menos Vendida": df["Categoria do Produto"].value_counts().idxmin()
    }

# Gerar a tabela com os indicadores das 4 lojas usando os DataFrames
tabela_indicadores = pd.DataFrame([
    analisar_loja(loja, "Loja 1"),
    analisar_loja(loja2, "Loja 2"),
    analisar_loja(loja3, "Loja 3"),
    analisar_loja(loja4, "Loja 4")
])

# Exibir a tabela ordenada por Faturamento
tabela_indicadores.sort_values(by="Faturamento Total (R$)", ascending=False)
```

Pontos fortes e fracos por loja:


*   Loja 1: Lidera em faturamento total e mantém
boa média de avaliação e diversidade de produtos vendidos. Entretanto, possui o maior custo médio de frete.
*   Loja 2: Segundo maior faturamento, com boa avaliação e variedade razoável de produtos. Também apresenta uma leve vantagem no frete médio em relação à Loja 1.
* Loja 3: Embora tenha o menor número de avaliações ruins e a melhor média de avaliação (4,05), seu faturamento é o segundo menor. Ainda assim, está próxima das líderes.
* Loja 4: Apresenta o menor faturamento total, menor custo médio de frete, e uma avaliação mediana (3,99). É também a única loja onde “instrumentos musicais” se destacam negativamente como categoria menos vendida, indicando baixa diversidade ou apelo comercial. Além disso, produtos como "Guitarra" figuram como os menos vendidos.

Conclusão

Com base nos indicadores analisados, a Loja 4 é a que apresenta o desempenho mais fraco entre as quatro unidades:

*   Tem o menor faturamento total, o que é um indicativo direto de menor retorno financeiro.
*   Embora tenha o frete médio mais barato, isso não compensa a menor atratividade comercial dos seus produtos, nem as vendas mais baixas.
* Sua avaliação média é apenas razoável, e inferior às das lojas 2 e 3.
